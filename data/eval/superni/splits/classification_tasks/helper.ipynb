## Top 20

```python
import pandas as pd

# Load task metadata
info = pd.read_csv("test_tasks_info.csv")

# Define complexity score (customize as you like)
info["complexity"] = info["# Answer Choices"] * info["# Instances"]

# Sort by complexity, descending
top20 = info.sort_values("complexity", ascending=False).head(20)

# Save task names only
top20["Name"].to_csv("test_tasks_top20.txt", index=False, header=False)

print("Saved top 20 tasks to test_tasks_top20.txt")
print("Tasks selected:", top20["Name"].tolist())
```

## Bottom 20

```python
import pandas as pd

# Load task metadata
info = pd.read_csv("test_tasks_info.csv")

# Define complexity score (customize as you like)
info["complexity"] = info["# Answer Choices"] * info["# Instances"]

# Sort by complexity, ascending this time
bottom20 = info.sort_values("complexity", ascending=True).head(20)

# Save task names only
bottom20["Name"].to_csv("test_tasks_bottom20.txt", index=False, header=False)

print("Saved bottom 20 tasks to test_tasks_bottom20.txt")
print("Tasks selected:", bottom20["Name"].tolist())
```
